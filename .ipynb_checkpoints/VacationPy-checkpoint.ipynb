```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
#Load the CSV Output File From Part 1 Into This File
file = "Weather_Output.csv"
weather_output_df = pd.read_csv(file)
weather_output_df
```

```python
#Create the Humidity Heatmap

#Configure the GMaps API Key
gmaps.configure(g_key)

#Plot the HeatMap
#Code Found from APIs Day 3 Activity 7
fig =gmaps.figure(map_type = "ROADMAP")

#Specify the Locations
locations = weather_output_df[['Lat','Lng']]

#Specify the Weights
humidity = weather_output_df['Humidity'].astype(float)

#Create the Heatmap Layer
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, dissipating = False, max_intensity = 10, point_radius = 3)

#Append the Heatmap Layer to the Heatmap
fig.add_layer(heat_layer)

#Display the Heatmap
fig
```

```python
#Create a DataFrame of Ideal Cities to Visit Based on Weather Criteria

#Define Desired Temperature Range
min_temp = 60
max_temp = 70

#Define the Desired Humidity Range
min_humid = 0
max_humid = 30

#Define the Desired Cloudiness Range
min_cloud = 20
max_cloud = 40

#Define the Desired Wind Speed Range
min_wind = 5
max_wind = 10

#Create the DataFrame of Ideal Cities
ideal_cities = weather_output_df.loc[(weather_output_df['Max Temp']>min_temp) &
                                     (weather_output_df['Max Temp']<max_temp) &
                                     (weather_output_df['Humidity']>min_humid) &
                                     (weather_output_df['Humidity']<max_humid) &
                                     (weather_output_df['Cloudiness']>min_cloud) &
                                     (weather_output_df['Cloudiness']<max_cloud) &
                                     (weather_output_df['Wind Speed']>min_wind) &
                                     (weather_output_df['Wind Speed']<max_wind)]
ideal_cities = ideal_cities.reset_index(drop=True)
ideal_cities
```

```python
#Use the Google Places API to Find the First Hotel for Each City Located 5000 Meters of Your Coordinates

#Create an Empty DataFrame to Assign the Hotels
#ideal_cities['Hotels'] = ""

#Iterate Through Rows to Find the First Hotel for Each City Located 5000 Meters of The Coordinates
for index, rows in ideal_cities.iterrows():
    #Create a Location Parameter
    location = ideal_cities[['Lat', 'Lng']]
    #Create the URL
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {"type":"lodging", "key":g_key, "radius":5000, "location":location}
    url = requests.get(base_url, params = params)
    response = url.json()
    print(response)    
```
